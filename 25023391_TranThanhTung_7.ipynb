```python
#2
from math import *
n = int(input("Nhap so nguyen duong: "))
while n <= 0:
    n = int(input("Gia tri cua ban sai. Nhap lai: "))

for i in range(2, int(sqrt(n)) + 1):
    if n % i == 0:
        print("Day khong phai so nguyen to")
        break
else:
    print(n,": Day la so nguyen to")
```

```python
#W4A6
a, b = map(int, input().split())
s = 0
for i in range(a, b + 1):
    if i < 2: continue
    ok = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            ok = False
            break
    if ok: s += i
print(s)
```

```python
#W4A8
n = int(input())
c = 0
while True:
    s = str(n)
    if s == s[::-1]: break
    n += int(s[::-1])
    c += 1
print(c, n)
```

```python
#W4A9
n = int(input())
res = []
i = 1
while i*i <= n:
    sq = i*i
    s = str(sq)
    if len(set(s)) == len(s):
        res.append(sq)
    i += 1
if not res: print("(no number)")
else: print(*res)
```

```python
#W4A10
n = int(input())
m_len = 0
res = 1
for x in range(1, n + 1):
    tmp = x
    l = 1
    while tmp != 1:
        if tmp % 2 == 0: tmp //= 2
        else: tmp = 3 * tmp + 1
        l += 1
    if l > m_len:
        m_len = l
        res = x
print(res, m_len)
```

```python
#W4A11
n = int(input())
c = 0
for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        if i % 2 == 0: c += 1
        if i*i != n and (n//i) % 2 == 0: c += 1
print(c)
```

```python
#W4A13
a, b = map(int, input().split())
sa = 0
for i in range(1, a):
    if a % i == 0: sa += i
sb = 0
for i in range(1, b):
    if b % i == 0: sb += i
if sa == b and sb == a: print("true")
else: print("false")
```

```python
#W5A4
n = int(input("Nhap so nguyen n"))
def shh(n):
    if n <= 1:
        return False
    sod = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            sod += i
    return sod == n
print(f"6 là số hoàn hảo? {shh(6)}")
```

```python
#W5A8
def hamming(x: int, y: int):
    return bin(x ^ y).count("1")

print(hamming(1, 4))  
print(hamming(7, 10)) 
print(hamming(15, 0)) 
```

```python
#W5A10
def iso(a: str, b: str):
    if len(a) != len(b):
        return False
    
    m1 = {}  
    m2 = {}  
    
    for x, y in zip(a, b):
        if x in m1:
            if m1[x] != y:
                return False
        else:
            m1[x] = y
        
        if y in m2:
            if m2[y] != x:
                return False
        else:
            m2[y] = x
    
    return True

print(iso("abca", "zbxz"))  
print(iso("foo", "bar"))  
```

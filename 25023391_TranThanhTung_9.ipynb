```python
#W6A4
def solve():
    s = input("Nhập chuỗi các cặp key:value: ")
    pairs = s.split()
    result_dict = {}
    for pair in pairs:
        if ":" in pair:
            key, value = pair.split(":")
            if key not in result_dict:
                result_dict[key] = []
            result_dict[key].append(value)
    print(result_dict)

solve()
```

```python
#W6A9
def gop(d1, d2):
    ket_qua = d1.copy()

    for key, value in d2.items():
        if key in ket_qua:
            ket_qua[key] += value
        else:
            ket_qua[key] = value
    
    a = dict(sorted(ket_qua.items()))
    
    return a

dict1 = {'a': 100, 'b': 200, 'c': 300}
dict2 = {'a': 300, 'b': 200, 'd': 400}

print("Dict 1:", dict1)
print("Dict 2:", dict2)

result = gop(dict1, dict2)
print(result)
```

```python
#W6A10
lst = [1, 2, 3, 4, 5]
k = 5
result = []
i = 0
while i < len(lst):
    j = i + 1
    while j < len(lst):
        if lst[i] + lst[j] == k:
            result.append((lst[i], lst[j]))
        j += 1
    i += 1

n = len(result)
for i in range(n):
    for j in range(0, n - i - 1):
        if result[j][0] > result[j + 1][0]:
            temp = result[j]
            result[j] = result[j + 1]
            result[j + 1] = temp
print(result)
```

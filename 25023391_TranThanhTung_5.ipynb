```python
#W3A16
so_thuc = float(input("Nhập vào một số thực: "))

phan_nguyen = int(so_thuc)
phan_thap_phan = so_thuc - phan_nguyen

if so_thuc >= 0:
    tron_xuong = phan_nguyen
else:
    tron_xuong = phan_nguyen - (phan_thap_phan != 0)

if so_thuc >= 0:
    tron_len = phan_nguyen + (phan_thap_phan != 0)
else:
    tron_len = phan_nguyen

if so_thuc >= 0:
    tron_gan_nhat = phan_nguyen + (phan_thap_phan >= 0.5)
else:
    tron_gan_nhat = phan_nguyen - (abs(phan_thap_phan) >= 0.5)

print(tron_xuong, tron_len, tron_gan_nhat)
```

```python
#W3A17
a = int(input('Nhập a: '))
b = int(input('Nhập b: '))
c = int(input('Nhập c: '))
d = int(input('Nhập d: '))

if a != 0 and b % a == 0:
    cong_boi = b // a  
    if b * cong_boi == c and c * cong_boi == d:
        print('4 số tạo thành cấp số nhân với công bội:', cong_boi)
    else:
        print('4 số không tạo thành cấp số nhân theo thứ tự a, b, c, d')
else:
    print('4 số không tạo thành cấp số nhân (b không chia hết cho a)')
```
